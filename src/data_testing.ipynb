```python
import pandas as pd
```

```python
df.rename(columns={
    "Anio": "Anio",
    "Idestado": "ID_Estado",
    "Nomestado": "Estado",
    "Idddr": "ID_DDR",
    "Nomddr": "DDR",
    "Idcader": "ID_CADER",
    "Nomcader": "CADER",
    "Idmunicipio": "ID_Municipio",
    "Nommunicipio": "Municipio",
    "Idciclo": "ID_Ciclo",
    "Nomcicloproductivo": "Ciclo_Producto",
    "Idmodalidad": "ID_Modalidad",
    "Nommodalidad": "Modalidad",
    "Idunidadmedida": "ID_Unidad_Medida",
    "Nomunidad": "Unidad_Medida",
    "Idcultivo": "ID_Cultivo",
    "Nomcultivo Sin Um": "Cultivo",
    "Sembrada": "Superficie_Sembrada",
    "Cosechada": "Superficie_Cosechada",
    "Siniestrada": "Superficie_Siniestrada",
    "Volumenproduccion": "Produccion_Volumen",
    "Rendimiento": "Rendimiento",
    "Precio": "Precio_Promedio",
    "Valorproduccion": "Valor_Produccion",
    "Nomcultivo": "Cultivo_Extra",
    "Preciomediorural": "Precio_Rural_Extra"
}, inplace=True)

# Combinar datos de 'Cultivo_Extra' con 'Cultivo'
df['Cultivo'] = df['Cultivo'].combine_first(df['Cultivo_Extra'])

# Combinar datos de 'Precio_Promedio' con 'Precio_Rural_Extra'
df['Precio_Promedio'] = df['Precio_Promedio'].combine_first(df['Precio_Rural_Extra'])

# Eliminar columnas redundantes después de la concatenación
df.drop(columns=["Cultivo_Extra", "Precio_Rural_Extra"], inplace=True)
```

```python
nulos_por_columna = df.isnull().sum()
print(nulos_por_columna)
```

```python
muestras_aleatorias = df.sample(n=10, random_state=42)  # Puedes cambiar random_state para obtener diferentes muestras
print(muestras_aleatorias)
```

```python
import matplotlib.pyplot as plt

```

```python

# Filtrar datos para Jalisco
df_jalisco = df[df['Estado'] == 'Jalisco']

# Agrupar por cultivo, sumando el valor total de producción para identificar los 10 principales cultivos
top_10_cultivos = df_jalisco.groupby('Cultivo')['Valor_Produccion'].sum().nlargest(10).index

# Filtrar el DataFrame para incluir solo los cultivos principales
df_jalisco_top_10 = df_jalisco[df_jalisco['Cultivo'].isin(top_10_cultivos)]

# Agrupar por año y cultivo, sumando el valor de producción
jalisco_top_10_valor_produccion = df_jalisco_top_10.groupby(['Anio', 'Cultivo'])['Valor_Produccion'].sum().unstack()

# Crear la gráfica de tendencia para los 10 cultivos principales
plt.figure(figsize=(14, 8))
jalisco_top_10_valor_produccion.plot(ax=plt.gca(), linewidth=2)
plt.title("Tendencia del Valor de Producción en Jalisco - Top 10 Cultivos", fontsize=16)
plt.xlabel("Año", fontsize=14)
plt.ylabel("Valor de Producción (MXN)", fontsize=14)
plt.legend(title="Cultivos", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Convertir 'Rendimiento' a tipo numérico
df['Rendimiento'] = pd.to_numeric(df['Rendimiento'], errors='coerce')

# Eliminar valores faltantes en 'Rendimiento'
df = df.dropna(subset=['Rendimiento'])

# Filtrar datos para Jalisco
df_jalisco = df[df['Estado'] == 'Jalisco']

# Calcular los 10 cultivos con mayor rendimiento
top_10_rendimiento_cultivos = df_jalisco.groupby('Cultivo')['Rendimiento'].mean().nlargest(10).index

# Filtrar datos para los cultivos principales por rendimiento
df_jalisco_top_10_rendimiento = df_jalisco[df_jalisco['Cultivo'].isin(top_10_rendimiento_cultivos)]

# Agrupar por año y cultivo, sumando el valor de producción
jalisco_top_10_rendimiento = df_jalisco_top_10_rendimiento.groupby(['Anio', 'Cultivo'])['Valor_Produccion'].sum().unstack()

# Crear la gráfica de tendencia
plt.figure(figsize=(14, 8))
jalisco_top_10_rendimiento.plot(ax=plt.gca(), linewidth=2)
plt.title("Tendencia del Valor de Producción en Jalisco - Top 10 Cultivos por Rendimiento", fontsize=16)
plt.xlabel("Año", fontsize=14)
plt.ylabel("Valor de Producción (MXN)", fontsize=14)
plt.legend(title="Cultivos", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convertir 'Valor_Produccion' a numérico en caso de que no lo sea
df['Valor_Produccion'] = pd.to_numeric(df['Valor_Produccion'], errors='coerce')

# Eliminar valores faltantes en 'Valor_Produccion'
df = df.dropna(subset=['Valor_Produccion'])

# Agrupar por cultivo, sumando el valor total de producción para identificar los 10 principales cultivos
top_10_cultivos = df.groupby('Cultivo')['Valor_Produccion'].sum().nlargest(10).index

# Filtrar el DataFrame para incluir solo los cultivos principales
df_top_10_cultivos = df[df['Cultivo'].isin(top_10_cultivos)]

# Agrupar por año y cultivo, sumando el valor de producción
mexico_top_10_valor_produccion = df_top_10_cultivos.groupby(['Anio', 'Cultivo'])['Valor_Produccion'].sum().unstack()

# Crear la gráfica de tendencia
plt.figure(figsize=(14, 8))
mexico_top_10_valor_produccion.plot(ax=plt.gca(), linewidth=2)
plt.title("Tendencia del Valor de Producción en México - Top 10 Cultivos", fontsize=16)
plt.xlabel("Año", fontsize=14)
plt.ylabel("Valor de Producción (MXN)", fontsize=14)
plt.legend(title="Cultivos", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Asegurarse de que las columnas relevantes sean numéricas
df['Valor_Produccion'] = pd.to_numeric(df['Valor_Produccion'], errors='coerce')
df['Superficie_Sembrada'] = pd.to_numeric(df['Superficie_Sembrada'], errors='coerce')

# Eliminar valores faltantes en las columnas relevantes
# df = df.dropna(subset=['Valor_Produccion', 'Superficie_Sembrada'])∫

# Calcular el valor de producción por superficie sembrada
df['Valor_Por_Superficie'] = df['Valor_Produccion'] / df['Superficie_Sembrada']

# Agrupar por cultivo para obtener el promedio del valor por superficie
cultivos_valor_superficie = df.groupby('Cultivo')['Valor_Por_Superficie'].mean().nlargest(15)

# Crear la gráfica de los 15 cultivos con mejor valor de producción por superficie sembrada
plt.figure(figsize=(14, 8))
cultivos_valor_superficie.sort_values().plot(kind='barh', color='skyblue')
plt.title("Top 15 Cultivos en México por Valor de Producción por Superficie Sembrada", fontsize=16)
plt.xlabel("Valor de Producción por Hectárea (MXN)", fontsize=14)
plt.ylabel("Cultivo", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Asegurarse de que las columnas relevantes sean numéricas
df['Valor_Produccion'] = pd.to_numeric(df['Valor_Produccion'], errors='coerce')
df['Superficie_Sembrada'] = pd.to_numeric(df['Superficie_Sembrada'], errors='coerce')

# Eliminar valores faltantes en las columnas relevantes
df = df.dropna(subset=['Valor_Produccion', 'Superficie_Sembrada'])

# Calcular el valor de producción por superficie sembrada
df['Valor_Por_Superficie'] = df['Valor_Produccion'] / df['Superficie_Sembrada']

# Agrupar por cultivo para obtener el promedio del valor por superficie
cultivos_valor_superficie = df.groupby('Cultivo')['Valor_Por_Superficie'].mean().nlargest(15)

# Crear la gráfica de los 15 cultivos con mejor valor de producción por superficie sembrada
plt.figure(figsize=(14, 8))
cultivos_valor_superficie.sort_values().plot(kind='barh', color='skyblue')
plt.title("Top 15 Cultivos en México por Valor de Producción por Superficie Sembrada", fontsize=16)
plt.xlabel("Valor de Producción por Hectárea (MXN)", fontsize=14)
plt.ylabel("Cultivo", fontsize=14)

# Formatear el eje x para que no use notación científica
plt.gca().xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
