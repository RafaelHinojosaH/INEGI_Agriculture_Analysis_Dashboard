```python
def load_and_print_info(url):
    try:
        # Intentar cargar el dataset con codificación UTF-8
        df = pd.read_csv(url, encoding='utf-8')
    except UnicodeDecodeError:
        # Si falla, intentar con codificación ISO-8859-1
        df = pd.read_csv(url, encoding='ISO-8859-1')

    # Imprimir los campos y sus tipos
    print(f"Información del dataset: {url}")
    print(df.dtypes)
    print("\n")

    return df
```

```python
load_and_print_info(file_path)
```

```python
plt.style.use('custom_style')

# Carga del archivo CSV
def plot_graphs(file_path):
    df = pd.read_csv(file_path, encoding="utf-8")

    # Lista de gráficos a generar
    charts = [
        ("Cultivo", "Superficie_Sembrada", "Superficie Sembrada por Cultivo"),
        ("Cultivo", "Superficie_Cosechada", "Superficie Cosechada por Cultivo"),
        ("Cultivo", "Produccion_Volumen", "Volumen de Producción por Cultivo"),
        ("Cultivo", "Rendimiento", "Rendimiento por Cultivo"),
        ("Cultivo", "Precio_Promedio", "Precio Promedio por Cultivo"),
        ("Ano", "Superficie_Sembrada", "Superficie Sembrada por Año"),
        ("Ano", "Produccion_Volumen", "Volumen de Producción por Año"),
        ("Ano", "Rendimiento", "Rendimiento Promedio por Año"),
        ("Estado", "Superficie_Sembrada", "Superficie Sembrada por Estado"),
        ("Estado", "Valor_Produccion", "Valor de Producción por Estado"),
    ]

    # Convertir columnas numéricas si es necesario
    numeric_columns = ["Superficie_Sembrada", "Superficie_Cosechada", "Produccion_Volumen"]
    for col in numeric_columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

    # Generar los gráficos usando los 10 valores mayores
    for x, y, title in charts:
        plt.figure(figsize=(10, 6))
        top_10 = df.groupby(x)[y].sum().nlargest(10)
        top_10.plot(kind="bar", title=title)
        plt.xlabel(x)
        plt.ylabel(y)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

# Llamada a la función con el archivo CSV
file_path = "../data/transformed_combined_data.csv"  # Reemplaza con la ruta a tu archivo CSV
plot_graphs(file_path)
```
